# Basic Statistics
# To perform descriptive analytics, visualize data distributions, and preprocess the dataset for further analysis.
## Q.1 Descriptive Analytics for Numerical Columns
### To compute and analyse basic statistical measures for numerical columns in the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("https://raw.githubusercontent.com/SayaliD6/Datasets_used_in_assignments/main/BS1_sales_data_with_discounts.csv")
```

```python
df
```

```python
df.dtypes
```

```python
df.describe()
```

```python
numerical_columns = df.select_dtypes(include=['int', 'float']).columns
for column_name in numerical_columns:
    # Calculate mean
    mean_value = df[column_name].mean()

    # Calculate mode (mode returns multiple values if there are multiple modes)
    mode_values = df[column_name].mode()

    # Calculate median
    median_value = df[column_name].median()

    print("Column:", column_name)
    print("Mean:",mean_value)
    print("Mode:", mode_values.tolist())  # Convert mode values to list for printing
    print("Median:", median_value)
    print("\n")
```

## Q.2 Data Visualization
### To visualize the distribution and relationship of numerical and categorical variables in the dataset.

```python
numerical_columns = df.select_dtypes(include=['int', 'float']).columns

# Plot histograms for each numerical column
for column in numerical_columns:
    plt.hist(df[column], bins=10, alpha=0.5, label=column)
    plt.title('Histograms of Numerical Columns')
    # plt.xlabel('Value')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
for column in numerical_columns:
    plt.figure()
    plt.boxplot(df[column])
    plt.xlabel(column)
    plt.ylabel('Value')
    plt.title('Box plot of ' + column)
    plt.show()
```

```python
categorical_columns = df[['Brand','BU']]

for column in categorical_columns:
    category_counts = df[column].value_counts()

    plt.figure(figsize=(8, 6))
    category_counts.plot(kind='bar', color='skyblue')
    plt.title('Frequency of ' + column)
    plt.xlabel('Categories')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.show()
```

# Standardization of Numerical Variables
## Q3 To scale numerical variables for uniformity, improving the datasetâ€™s suitability for analytical models.

```python
df[numerical_columns]
```

```python
# Standardize numerical columns
for column in numerical_columns:
    mu = df[column].mean()
    sigma = df[column].std()
    df[column] = (df[column] - mu) / sigma
```

# Conversion of Categorical Data into Dummy Variables
### Q.4 To transform categorical variables into a format that can be provided to ML algorithms.


```python
df.drop(columns=['Date','Day','City','SKU','Model','Brand'], inplace=True)
categorical_columns = df.select_dtypes(include=['object']).columns
df= pd.get_dummies(df, columns=categorical_columns)
print(df.head())
```
